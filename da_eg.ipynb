```python
# E.g.1 - Paper's Eg

print("E.g.1 - Paper's Eg: \n")

firms_pref = {
                        1: [[1,2],[1,5],[2,5],[1,3],[4,5],[2,4],[1,4],[3,4],[3,5],[2,3],[1],[4],[3],[2],[5]],
                        2: [[3,6],[3,5],[5,6],[2,5],[1,3],[2,6],[1,5],[1,2],[2,3],[1,6],[1],[2],[3],[5],[6]],
                        3: [[2,4],[1,2],[3,4],[2,3],[1,3],[1,4],[1],[2],[3],[4]]
              }

workers_pref = {
                        1: [[3],[1],[2]],
                        2: [[2,3],[1,3],[1,2],[1],[2],[3]],
                        3: [[1],[2]],
                        4: [[1],[3],[2]],
                        5: [[2],[3]],
                        6: [[1,3],[3],[1]]
                }
# 1. firms proposing:
m, f = defer_acceptance(males_prefs=firms_pref, females_prefs=workers_pref)
print(f'Firms proposing result:\n Male Matches: {m} \n Female Matches: {f}' )

print()

# 2. workers proposing:
m, f = defer_acceptance(males_prefs=workers_pref, females_prefs=firms_pref)
print(f'Firms proposing result:\n Male Matches: {m} \n Female Matches: {f}' )
```
